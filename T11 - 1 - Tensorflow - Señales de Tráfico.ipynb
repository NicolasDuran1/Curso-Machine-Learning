```python
#Debemos hacer bucles por cada una de las carpetas y en ellas buscar cada archivo

def load_ml_data(data_directory):
    directories = [d for d in os.listdir(data_directory)
                  if os.path.isdir(os.path.join(data_directory, d))]
    #print(directories)
    labels = []
    images = []
    for d in directories:
        label_dir = os.path.join(data_directory, d)
        file_names = [os.path.join(label_dir, f)
                     for f in os.listdir(label_dir)
                     if f.endswith(".ppm")]
        #print("\n",label_dir)
        #print("\n",file_names[0])
        
        for f in file_names:
            images.append(imd.imread(f))
            labels.append(int(d))
        
    return images, labels
```

```python
main_dir = "../datasets/belgian/"
train_data_dir = os.path.join(main_dir, "Training")
test_data_dir = os.path.join(main_dir, "Testing")
```

* Hay 62 clases de imágenes (desde la 0 hasta la 61).
* La distribución de señales de tráfico no es uniforme (algunas salen más veces que otras).

```python
from skimage import transform
```

```python
# Quiero quedarme con la altura y anchura más pequeña

w = 9999
h = 9999 #weigth y heigth
for image in images:
    if image.shape[0] < h:
        h = image.shape[0] #el shape de una imagen se representa por altura, anchura y canales de color. Shape[0] corresponde a la altura de esa imagen.
    if image.shape[1] < w:
        w = image.shape[1]
print("Tamaño mínimo: {0}x{1}".format(h,w))
```

# C213: Creación del modelo de la Red Neuronal

```python
#Inicializar un objeto de tipo grafo (graph)
x = tf.placeholder(dtype = tf.float32, shape = [None, 30, 30]) #Es como decirle a tensorflow que me guarde espacio para guardar objetos de 30x30
y = tf.placeholder(dtype = tf.int32, shape = [None])

#Construir la red neuronal
images_flat = tf.contrib.layers.flatten(x) #Será distribuir los datos de matrices de 30x30 a una lista plana

#capa de conexion estandar
logits = tf.contrib.layers.fully_connected(images_flat, 62, tf.nn.relu) 
#entrada de la red neuronal images_flat, 62 posibles clases de salida, relu llevar a cabo la clasificación

#Función de pérdida abajo
loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels = y, logits = logits)) #logits son los creados anteriormente como paraetros de entrada
#Cada entrada (señal de transito) solo puede tener una sola clase de salida.

#Función de optimización: depende del algoritmo que utilicemos. Distintos optimizadores.
train_opt = tf.train.AdamOptimizer(learning_rate = 0.001).minimize(loss)

final_pred = tf.argmax(logits, 1) #1 solo elemento, etiquetas con las cuales se clasificará

accuracy = tf.reduce_mean(tf.cast(final_pred, tf.float32))
```

# C214: Entrenamiento del modelo

```python
#Establecer una semilla para ser concistentes y no obtener resultados diferentes cada vez que se ejecuta la red.
tf.set_random_seed(1234)

sess = tf.Session()

sess.run(tf.global_variables_initializer())

for i in range(601):
    #feed_dict: parámetros que se suministran a la red neuronal.
    _, accuracy_val = sess.run([train_opt, accuracy],
                                 feed_dict = {
                                     x: images30,
                                     y: list(labels)
                                 }) #No me interesa el optimizador en este caso, solo la precisión del modelo.
    
    #_, loss_val = sess.run([train_opt, loss],
                             #feed_dict = {
                                 #x: images30,
                                 #y: list(labels)
                             #})
    if i%50 == 0:
        print("EPOCH", i)
        print("Eficacia: ", accuracy_val) #Si quiero la pérdida debo cambiar lo que se imprime aquí y en sess.run
        #print("Pérdidas: ", loss_val)
    #print("Fin del EPOCH ", i)
```

# C215: Validación del modelo / Evaluación de la red neuronal

```python
sample_idx = random.sample(range(len(images30)), 40)
sample_images = [images30[i] for i in sample_idx]
sample_labels = [labels[i] for i in sample_idx]
```

```python
prediction = sess.run([final_pred], feed_dict={x:sample_images})[0]
```

```python
#Prediccion de la red neuronal
prediction
```

```python
#Clasificacion original de las imagenes. Comparar con la prediccion

sample_labels
```

```python
plt.figure(figsize=(16,20))
for i in range(len(sample_images)):
    truth = sample_labels[i]
    predi = prediction[i]
    plt.subplot(10,4,i+1)
    plt.axis("off")
    color = "green" if truth==predi else "red"
    plt.text(40,15, "Real:         {0}\nPredicción:{1}".format(truth, predi), fontsize = 14, color = color)
    plt.imshow(sample_images[i], cmap="gray")
plt.show()
```
