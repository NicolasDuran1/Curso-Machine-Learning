# C219: Tensorflow y la regresión softmax

### y = softmax(W * x + b)

```python
dimension_input = 784
n_categories = 10
```

```python
x = tf.placeholder(tf.float32, [None, dimension_input]) #None: la dimension en principio puede ser de cualquier longitud.
```

```python
W = tf.Variable(tf.zeros([dimension_input,n_categories])) #Matriz de ceros de 784 filas y 10 columnas
b = tf.Variable(tf.zeros([n_categories])) #Vector de 10 columnas
```

```python
softmax_args = tf.matmul(x,W) + b #matmul: multiplicación de matrices. Traspuesta

#Predicción Y TONGO: Y HAT
y_hat = tf.nn.softmax(softmax_args) #nn: neuronal networks
```

# C120: Entrenando la red neuronal

* Loss/Cost <- objetivo minimizar las pérdidas

```python
from IPython.display import display, Math, Latex
```

```python
display(Math(r"H_{y}(\hat{y}) = -\sum_{i} y_i log(\hat{y_i})"))
```

```python
y_ = tf.placeholder(tf.float32, [None, 10])
```

```python
#tf.reduce_mean: normaliza los datos
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y_hat), reduction_indices = [1])) # Sumatoria de la función
```

#### También se puede aplicar de forma automática y no desglozando la fórmula mediante una función de Tensorflow

```python
#tf.nn.softmax_cross_entropy_with_logits(softmax_args, y_) #Ejecutar si no queremos aplicar la función de arriba en "cross_entropy"
```

```python
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy) #Cómo será entrenada la red neuronal
```

```python
session = tf.InteractiveSession()
```

```python
tf.global_variables_initializer().run()
```

```python
#Se reduce el costo de memoria y computacional ya que no entrena con todo el conjunto de imágenes al mismo tiempo

for _ in range(10000): #Realiza el entrenamiento 10000 veces pero con 150 muestras cada vez. Es un entrenamiento "Estocástico"
    batch_x, batch_y = mnist.train.next_batch(150) #Toma 150 imagenes al azar cada vez
    session.run(train_step, feed_dict = {x:batch_x, y_:batch_y})
```

# C221: Evaluación de la red neuronal

```python
correct_predictions = tf.equal(tf.argmax(y_hat, 1), tf.argmax(y_,1))
```

```python
accuracy = tf.reduce_mean(tf.cast(correct_predictions, tf.float32))
```

```python
print("Precisión del conjunto de Test: ", session.run(accuracy, feed_dict = {x:mnist.test.images, y_:mnist.test.labels}))
```
