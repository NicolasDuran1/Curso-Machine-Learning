# C51: Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gender = ["Male", "Female"];
income = ["Poor", "Middle Class", "Rich"];
```

```python
n = 500; #Filas

gender_data = [];
income_data = [];

for i in range(0, n):
    gender_data.append(np.random.choice(gender));
    income_data.append(np.random.choice(income));
```

```python
gender_data[:5] #Mostrar 5 datos de los 500.
```

```python
income_data[:5]
```

```python
# Z -> N(0,1)
# N(m, s) -> m + s * Z

height = 160 + 30 * np.random.randn(n); #media = 160 (cm), desviacion estandar = 30, randn = random normal
weight = 65 + 25 * np.random.randn(n);
age = 30 + 12 * np.random.randn(n); #12 años arriba, 12 años abajo
income = 18000 + 3500 * np.random.rand(n);
```

```python
age = (int(i) for i in age) #Transformar cada celda de la lista en entero

data = pd.DataFrame (
    {
        "Gender": gender_data,
        "Economic Status": income_data,
        "Height": height,
        "Weight": weight,
        "Age": age,
        "Income": income
    }
)
```

```python
data.head()
```

# C52: Agrupación de los datos por categoría

```python
grouped_gender = data.groupby("Gender") #Objecto resultante. Divide en dos, en el nombre y el grupo del objecto
```

```python
grouped_gender.groups
```

### Mostrar los grupos contenidos en el objeto

```python
for names, groups in grouped_gender:
    print(names); #Contiene el nombre del grupo, en este caso se han agrupado por Female y Male
    print(groups);#Contiene toda la información contenida en un solo grupo (Female o Male).
    
    #Recordar que el método groupby genera un objeto que contiene dos contenidos, nombre y grupo.
```

### Mostrar un solo grupo

```python
grouped_gender.get_group("Male")[:10]
```

### Agrupar más de una categoría

```python
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group) #genera 6 grupos, 3 para cada Género. 
#Habrán Grupos de Mujeres: Pobres, Clase Media y Ricas.
#Habrán Grupos de Hombres: Pobres, Clase Media y Ricas.
```

```python
double_group.get_group(("Male", "Rich"))[:10]
```

```python
for name, groups in double_group:
    print (name)
    print (groups)
```

## Operaciones sobre datos agrupados

```python
double_group.sum()
```

```python
double_group.mean()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
grouped_income = double_group["Income"];
```

```python
grouped_income.describe()
```

```python
#El método aggregate es una modificacion, de cierta forma, del groupby

double_group.aggregate(
    {
        "Income": np.sum,
        "Age": np.mean,
        "Height": np.std
    
    }
)
```

```python
double_group.aggregate(
    {
        "Age": np.mean,
        "Height": lambda h: np.mean(h)/np.std(h) #Dividir el promedio de las alturas por su desviacion tipica 
        #Se declara la variable h con un valor con la funcion lambda
    }
)
```

### Aplicar funciones a todas las variables del grupo

```python
double_group.aggregate([np.sum, np.mean,np.std]) #std = standar deviation
```

```python
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```

## Transformación de variables

La transformación de variables nos permite que a todos o algunos de los datos presentes en el data set se aplica alguna transformación o normalización aritmética o de cambio de valores, en donde se aplica lo que se defina en los parámetros de la función *transform*.

```python
#Sintaxis-> lambda arguments : expression
#Lambda permite crear una función en donde la expresión definida determina el valor de la variable dada

zscore = lambda x: (x - x.mean()) / x.std();
```

```python
z_groupe = double_group.transform(zscore)
```

```python
z_groupe[:10]
```

```python
plt.hist(z_groupe["Age"])
```

### Rellenar valores nulos con la función Transform

```python
fill_na_mean = lambda x: x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)[:10]
```
